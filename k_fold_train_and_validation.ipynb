```python
from k_fold import run_k_fold, run_k_fold_pretrained
```

```python
epochs, k = 20, 5
```

```python
all_results = run_k_fold(epochs, k)
```

```python
for i, result in enumerate(all_results):
    loss = result.sort_values(by='validation_loss')['validation_loss'].values[0]
    acc = result.sort_values(by='validation_accuracy', ascending=False)['validation_accuracy'].values[0]
    
    print('{} fold: loss={}, acc={}'.format(i + 1, loss, acc))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

for i, result in enumerate(all_results):
    plt.plot(np.arange(0, epochs) + 1, result['validation_loss'], color=colors[i], label='Validation loss K={}'.format(i + 1))
    plt.plot(np.arange(0, epochs) + 1, result['train_loss'], color=colors[i], label='Train loss K={}'.format(i + 1), linestyle='--')
    
plt.title('5-Fold cross validation')
plt.legend()
plt.xlabel('Epochs', fontsize=11)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

for i, result in enumerate(all_results):
    plt.plot(np.arange(0, epochs) + 1, result['validation_accuracy'], color=colors[i], label='Validation accuracy K={}'.format(i + 1))
    plt.plot(np.arange(0, epochs) + 1, result['train_accuracy'], color=colors[i], label='Train accuracy K={}'.format(i + 1), linestyle='--')
    
plt.title('5-Fold cross validation')
plt.legend()
plt.xlabel('Epochs', fontsize=11)
plt.show()
```
