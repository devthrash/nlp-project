Cache the encoder dictionary on disk, so we don't have to read the entire dataset each time when we want to encode the words.

```python
from ironydetection.preprocessor import DataframePreprocessor, load_csv
import os
import numpy as np

preprocessor = DataframePreprocessor()
df = load_csv(os.path.join('input', 'datasets', 'train', 'SemEval2018-T3-train-taskA_emoji_ironyHashtags.txt'))
df = preprocessor.preprocess(df)

vocab = np.concatenate(df['tweet'].values)
```

```python
vocab.tolist()
```

```python
from ironydetection.net.encoder import Encoder
```

```python
encoder = Encoder(vocab.tolist(), 'unknown')

encoder.encode('sweet')
```

```python
encoder._dictionary
```

```python
import pickle

with open(os.path.join('output', 'custom_encoder.pickle'), 'wb') as handle:
    pickle.dump(encoder, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
